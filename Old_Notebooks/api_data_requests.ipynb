```python
# Dependencies and Setup
import pandas as pd
from sodapy import Socrata
import time

# Import API key
from config import app_token
from config import api_key_id
from config import api_secret_key
```

```python
start = time.perf_counter()

# create client variable with parameters
client = Socrata('data.austintexas.gov',
                 app_token=app_token,
                 username=api_key_id,
                 password=api_secret_key)

# retrieve data; look into limit ?get_all?
results = client.get("fdj4-gpfu", limit=3000000)

# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)

end = time.perf_counter()
print(f'{end-start}')
```
