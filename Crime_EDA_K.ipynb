```python
#import necessary dependencies 
import pandas as pd
import numpy as np
```

```python
#read in csv 
data = pd.read_csv('Crime_Reports.csv')
```

```python
#how many unique values are there in highest offense description?
data['Highest Offense Description'].nunique()
```

```python
#convert occurred date time to datetime dtype
data['Occurred Date Time'] =  pd.to_datetime(data['Occurred Date Time'])
```

```python
#create a year column for occurred date time 
data['year'] = pd.DatetimeIndex(data['Occurred Date Time']).year
```

```python
#filter dataframe to be between 2018-2021
data = data[data['year']>=2018]
data = data[data['year']<=2021]
```

```python
#check value counts for each column
data['Highest Offense Description'].value_counts()
```

```python
#check the highest offense description for filtered df 
data['Highest Offense Description'].unique()
```

```python
# check unique zip codes
#we can consider dropping the 786 zip codes (typically outer cities)
data['Zip Code'].unique()
```

```python
#check null values
#do we want to drop null values or do we want to drop the columns? 
data.isnull().sum(axis = 0)
```

```python
#removing spaces in column names 
data.columns = data.columns.str.replace(' ','_')
```

```python
#preview location type to see if its beneficial
data["Location_Type"].sample(7)
```

```python
#preview PRA to see if its beneficial
data["PRA"].sample(7)
```

```python
#preview clearance date to see if its beneficial
data["Clearance_Date"].sample(10)
#data["Clearance_Date"].dtype
```

```python
#drop unneccessary columns for our analysis 
clean_df= data.drop(["Council_District","APD_Sector","APD_District","PRA"],axis = 1)
```

```python
#dropping nulls of important columns 
clean_df = clean_df.dropna(axis=0, subset=["Location_Type","Zip_Code","Census_Tract",
                                          "Latitude"])
```

```python
#replacing nulls for UCR_Category with zero 
clean_df[["UCR_Category"]] = clean_df[["UCR_Category"]].fillna('0')
```

```python
#replacing nulls with "None"
clean_df[["Category_Description"]] = clean_df[["Category_Description"]].fillna('None')
```

```python
#replacing nulls with "U" for unkown 
clean_df[["Clearance_Status"]] = clean_df[["Clearance_Status"]].fillna('U')
```

```python
#replacing nulls with a random date so its obvious on visualizaitons 
clean_df[["Clearance_Date"]] = clean_df[["Clearance_Date"]].fillna('01/01/1888')
```

```python
#check nulls
clean_df.isnull().sum(axis = 0)
```

```python
#check dtypes
clean_df.dtypes
```

```python
#change to datetime: Occurred_Date, Occurred_Time, Report_Date_Time, Report_Date, 
##Report_Time
clean_df["Occurred_Date"] =  pd.to_datetime(clean_df["Occurred_Date"])
clean_df["Occurred_Time"] =  pd.to_datetime(clean_df["Occurred_Time"])
clean_df["Report_Date_Time"] =  pd.to_datetime(clean_df["Report_Date_Time"])
clean_df["Report_Date"] =  pd.to_datetime(clean_df["Report_Date"])
clean_df["Report_Time"] =  pd.to_datetime(clean_df["Report_Time"])
```

```python
#check dtypes
clean_df.dtypes
```

```python
#convert clean_df to csv 
data.to_csv(r'Resources/2018To2021CrimeData.csv')
```
